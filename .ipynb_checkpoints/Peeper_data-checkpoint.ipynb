## PCA

```python
pca = PCA(n_components=2)
scaled = StandardScaler().fit_transform(df[to_plot])
pca_df = pd.DataFrame(pca.fit_transform(scaled), columns=['PC1', 'PC2'])
pca_df['month'] = df.month
pca_df['depth'] = df.Depth
pca_df['depth_group'] = df.Depth_group
pca_df.head()
```

```python
def confidence_ellipse(x, y, ax, n_std=1.5, facecolor='none', **kwargs):
    """
    Source: https://matplotlib.org/stable/gallery/statistics/confidence_ellipse.html
    Create a plot of the covariance confidence ellipse of *x* and *y*.

    Parameters
    ----------
    x, y : array-like, shape (n, )
        Input data.

    ax : matplotlib.axes.Axes
        The axes object to draw the ellipse into.

    n_std : float
        The number of standard deviations to determine the ellipse's radiuses.

    **kwargs
        Forwarded to `~matplotlib.patches.Ellipse`

    Returns
    -------
    matplotlib.patches.Ellipse
    """
    if x.size != y.size:
        raise ValueError("x and y must be the same size")

    cov = np.cov(x, y)
    pearson = cov[0, 1]/np.sqrt(cov[0, 0] * cov[1, 1])
    # Using a special case to obtain the eigenvalues of this
    # two-dimensional dataset.
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0), width=ell_radius_x * 2, height=ell_radius_y * 2,
                      facecolor=facecolor, **kwargs)

    # Calculating the standard deviation of x from
    # the squareroot of the variance and multiplying
    # with the given number of standard deviations.
    scale_x = np.sqrt(cov[0, 0]) * n_std
    mean_x = np.mean(x)

    # calculating the standard deviation of y ...
    scale_y = np.sqrt(cov[1, 1]) * n_std
    mean_y = np.mean(y)

    transf = transforms.Affine2D() \
        .rotate_deg(45) \
        .scale(scale_x, scale_y) \
        .translate(mean_x, mean_y)

    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)
```

```python
fig, ax = plt.subplots(1,1, figsize=(7, 4))

confidence_ellipse(pca_df[pca_df.depth_group=='Top']['PC1'], pca_df[pca_df.depth_group=='Top']['PC2'], ax, edgecolor='tab:blue', linestyle='--', alpha=0.9)
confidence_ellipse(pca_df[pca_df.depth_group=='Mid']['PC1'], pca_df[pca_df.depth_group=='Mid']['PC2'], ax, edgecolor='tab:orange', linestyle='--', alpha=0.9)
confidence_ellipse(pca_df[pca_df.depth_group=='Btm']['PC1'], pca_df[pca_df.depth_group=='Btm']['PC2'], ax, edgecolor='tab:green', linestyle='--', alpha=0.9)
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='depth_group', style='month', ax=ax, s=50, edgecolor='black', alpha=1.)

ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['top'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)

first_legend = ax.legend(handles = ax.get_legend_handles_labels()[0][4:],
          labels=['Sample Month', 'Oct.', 'Feb.', 'May', 'Aug.'],
          fontsize=12, title_fontsize=12, edgecolor='black', bbox_to_anchor=(1.05, .45))
ax.add_artist(first_legend)
ax.legend(handles = ax.get_legend_handles_labels()[0][0:4],
          labels=["Sample Depth", 'Top (0-20cm)', 'Middle (20-40cm)', 'Bottom (40-60cm)'],
          fontsize=12, title_fontsize=12, edgecolor='black', bbox_to_anchor=(1.05, 0.61))
plt.xlim([-3.6,6])
#plt.ylim([-3,3.5])
ax.tick_params(width=1.5)
plt.xticks(size=10)
plt.yticks(fontsize=10)
plt.xlabel('PC1 ({:.2f}%)'.format(pca.explained_variance_ratio_[0]*100), size=12)
plt.ylabel('PC2 ({:.2f}%)'.format(pca.explained_variance_ratio_[1]*100), size=12)
plt.tight_layout()
plt.savefig('figures/porewater/porewater_pca.png', dpi=450)
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(7, 4))

confidence_ellipse(pca_df[pca_df.month=='Oct']['PC1'], pca_df[pca_df.month=='Oct']['PC2'], ax, edgecolor='tab:red', linestyle='--', alpha=0.9)
confidence_ellipse(pca_df[pca_df.month=='Feb']['PC1'], pca_df[pca_df.month=='Feb']['PC2'], ax, edgecolor='tab:gray', linestyle='--', alpha=0.9)
confidence_ellipse(pca_df[pca_df.month=='May']['PC1'], pca_df[pca_df.month=='May']['PC2'], ax, edgecolor='tab:brown', linestyle='--', alpha=0.9)
confidence_ellipse(pca_df[pca_df.month=='Aug']['PC1'], pca_df[pca_df.month=='Aug']['PC2'], ax, edgecolor='tab:cyan', linestyle='--', alpha=0.9)


sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='month', style='depth_group', ax=ax, s=50, edgecolor='black', alpha=1.,palette=['tab:red','tab:gray','tab:brown','tab:cyan'])

ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['top'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)

first_legend = ax.legend(handles = ax.get_legend_handles_labels()[0][0:5],
          labels=['Sample Month', 'Oct.', 'Feb.', 'May', 'Aug.'],
          fontsize=12, title_fontsize=12, edgecolor='black', bbox_to_anchor=(1.05, .45))
ax.add_artist(first_legend)
ax.legend(handles = ax.get_legend_handles_labels()[0][5:],
          labels=["Sample Depth", 'Top (0-20cm)', 'Middle (20-40cm)', 'Bottom (40-60cm)'],
          fontsize=12, title_fontsize=12, edgecolor='black', bbox_to_anchor=(1.05, 0.61))
plt.xlim([-3.6,6])
#plt.ylim([-3,3.5])
ax.tick_params(width=1.5)
plt.xticks(size=10)
plt.yticks(fontsize=10)
plt.xlabel('PC1 ({:.2f}%)'.format(pca.explained_variance_ratio_[0]*100), size=12)
plt.ylabel('PC2 ({:.2f}%)'.format(pca.explained_variance_ratio_[1]*100), size=12)
plt.tight_layout()
plt.savefig('figures/porewater/porewater_pca_season_colored.png', dpi=450)
plt.show()
```

## Distance based analysis

```python
dist = pdist(scaled, metric='euclidean')
square_dist = squareform(dist)
dist_mat = distance.DistanceMatrix(square_dist, ids=pca_df.index)
```

```python
distance.permanova(dist_mat, grouping=list(pca_df.depth_group), permutations=99999)
```

```python
distance.permanova(dist_mat, grouping=list(pca_df.month), permutations=99999)
```

```python
distance.anosim(dist_mat, grouping=list(pca_df.depth_group), permutations=99999)
```

```python
distance.anosim(dist_mat, grouping=list(pca_df.month), permutations=99999)
```
