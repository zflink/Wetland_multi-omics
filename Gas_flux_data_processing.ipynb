```python
# Import necessary packages
import pandas as pd
from pandas.tseries.offsets import DateOffset
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as np
import os
from datetime import datetime, timedelta
from scipy.stats import linregress
from scipy.stats import normaltest, kruskal, f_oneway
from scikit_posthocs import posthoc_dunn
```

## Import data

```python
# Load in raw data from Picarro files
cols = ['DATE','TIME','N2O_dry','CH4_dry','CO2_dry','H2O','O2_dry','NH3','CH4','CO2']
df = pd.DataFrame()
for file in os.listdir("data/Gas_flux/"):
    if file.endswith('.dat'):
        temp_df = pd.read_table("data/Gas_flux/"+file, sep=r'\s+', usecols=cols)
        temp_df["DATE_TIME"] = pd.to_datetime(temp_df.DATE+" "+temp_df.TIME)
        if file == "JFAADS2178-20220520-121044-DataLog_User.dat":
            temp_df.DATE_TIME = temp_df.DATE_TIME - timedelta(hours=1) #Need to correct one file affected by instrument timechange    
        df = pd.concat([df, temp_df], ignore_index=True)
df.head()
```

```python
# Load in start and end times for each chamber experiment
times = pd.DataFrame()
for file in os.listdir("data/Gas_flux/"):
    if file.endswith('.txt'):
        times = pd.concat([times, pd.read_table("data/Gas_flux/"+file)], ignore_index=True)
                          
times.Start = pd.to_datetime(times.Start)
times.End = pd.to_datetime(times.End)
times['month_num'] = [i.month for i in times.Start]
times['month'] = times.month_num.map({10:'Oct', 2:'Feb', 5:'May', 8:'Aug'})
times.head()
```

## Compute slopes and plot concentration vs time

```python
def build_legend():
    #Build custom legend for concentration vs time plots
    red_patch = mpatches.Patch(color='tab:red', label='Oct')
    gray_patch = mpatches.Patch(color='tab:gray', label='Feb')
    brown_patch = mpatches.Patch(color='tab:brown', label='May')
    cyan_patch = mpatches.Patch(color='tab:cyan', label='Aug')
    plt.legend(handles=[red_patch, gray_patch, brown_patch, cyan_patch], loc='upper right')

colors = {'Oct':'tab:red','Feb':'tab:gray','May':'tab:brown','Aug':'tab:cyan'}
```

```python
# Plot CO2 vs time
pd.set_option('mode.chained_assignment', None)
CO2_slopes = []
for i in times.index:
    test = df[(df.DATE_TIME>times.iloc[i,2]+DateOffset(minutes=4))&(df.DATE_TIME<times.iloc[i,3]-DateOffset(minutes=1))]
    test["time_elapsed_s"] = (test.DATE_TIME - test.DATE_TIME.iloc[0])/timedelta(seconds=1)
    test['delta_CO2'] = test.CO2_dry - test.CO2_dry.iloc[0]
    plt.scatter(test.time_elapsed_s, test.delta_CO2, alpha=0.5, linewidth=0, s=2, label="Month "+str(times.iloc[i,5]), color=colors[times.iloc[i,5]])
    CO2_slopes.append(linregress(test.time_elapsed_s, test.CO2_dry).slope)
times["CO2_slopes"] = CO2_slopes
build_legend()
plt.title("CO$_2$ Concentration")
plt.ylabel("Change in CO$_2$ (ppm)")
plt.xlabel("Time (s)")
plt.tight_layout()
plt.savefig("figures/CO2_conc_v_time.png", dpi=450)
plt.show()
```

```python
# Plot CH4 vs time
CH4_slopes = []
for i in times.index:
    test = df[(df.DATE_TIME>times.iloc[i,2]+DateOffset(minutes=4))&(df.DATE_TIME<times.iloc[i,3]-DateOffset(minutes=1))]
    test["time_elapsed_s"] = (test.DATE_TIME - test.DATE_TIME.iloc[0])/timedelta(seconds=1)
    test["delta_CH4"] = test.CH4_dry - test.CH4_dry.iloc[0]
    plt.scatter(test.time_elapsed_s, test.delta_CH4, linewidth=0,alpha=0.9, s=1, label="Month "+str(times.iloc[i,5]), color=colors[times.iloc[i,5]])
    CH4_slopes.append(linregress(test.time_elapsed_s, test.CH4_dry).slope)
times["CH4_slopes"] = CH4_slopes
build_legend()
plt.title("CH$_4$ Concentration")
plt.ylabel("Change in CH$_4$ (ppm)")
plt.xlabel("Time (s)")
plt.tight_layout()
plt.savefig("figures/CH4_raw_17Aug2022.png", dpi=300)
plt.show()
```

## Convert slopes to fluxes

```python
# Ideal gas law calculations for each season
def FtoK(temp_F):
    """Convert temperature in F to Kelvin"""
    return (temp_F - 32)*5/9+273.15

def moles_gas(P, V, R, T):
    """Ideal gas law moles of gas"""
    return P*V/R/T

P = 1 #atm
V = 0.982 #L 0.8L Chamber + 0.1L (Picarro) + 0.082L (34ft tubing = 10.3632 m)
R = 0.08205746 #L atm K-1 mol-1

# Define different temperatures for each timepoint
T_oct = FtoK(57)
T_feb = FtoK(47)
T_may = FtoK(55)
T_aug = FtoK(81)

# Calculate moles of gas for each timepoint
n_oct = moles_gas(P, V, R, T_oct)
n_feb = moles_gas(P, V, R, T_feb)
n_may = moles_gas(P, V, R, T_may)
n_aug = moles_gas(P, V, R, T_aug)

n_dict = {'Oct':n_oct, 'Feb':n_feb, 'May':n_may, 'Aug':n_aug}
```

```python
#Chamber cross sectional area: 4" diameter = 10.16cm = 0.1016m --> pi*(d/2)^2 = 0.008
area = np.pi*(0.1016/2)**2 #m^2
times['n'] = times.month.map(n_dict)

times["CO2_flux"]= times.n*1e9*times.CO2_slopes/1e6/area #nmol/s/m^2
times["CH4_flux"] = times.n*1e9*times.CH4_slopes/1e6/area #nmol/s/m^2
times["N2O_flux"] = times.n*1e9*times.N2O_slopes/1e6/area #nmol/s/m^2
times.head()
```

## Statistics to compare glas flux across seasons

```python
print('CH4 normality test')
print(normaltest(times.CH4_flux))
print('CO2 normality test')
print(normaltest(times.CO2_flux))
print('N2O normality test')
print(normaltest(times.N2O_flux))
```

```python
kruskal(times[times.month=='Oct'].CH4_flux, times[times.month=='Feb'].CH4_flux, times[times.month=='May'].CH4_flux, times[times.month=='Aug'].CH4_flux)
```

```python
posthoc_dunn([times[times.month=='Oct'].CH4_flux, times[times.month=='Feb'].CH4_flux, times[times.month=='May'].CH4_flux, times[times.month=='Aug'].CH4_flux], p_adjust='fdr_bh')
```

**Significantly different CH4 fluxes across seaons. August is greater than Oct and Feb. May is greater than Oct and Feb. Feb is greater than Oct.**

```python
kruskal(times[times.month=='Oct'].CO2_flux, times[times.month=='Feb'].CO2_flux, times[times.month=='May'].CO2_flux, times[times.month=='Aug'].CO2_flux)
```

```python
posthoc_dunn([times[times.month=='Oct'].CO2_flux, times[times.month=='Feb'].CO2_flux, times[times.month=='May'].CO2_flux, times[times.month=='Aug'].CO2_flux], p_adjust='fdr_bh')
```

**Significantly different CO2 fluxes across seasons. May and August are less than Oct and Feb.**

```python
f_oneway(times[times.month=='Oct'].N2O_flux, times[times.month=='Feb'].N2O_flux, times[times.month=='May'].N2O_flux, times[times.month=='Aug'].N2O_flux)
```

**No significant difference in N2O flux across seasons**
